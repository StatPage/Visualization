###  점 그리기
점을 그리고 싶다 → matplotlib의 scatter method를 사용한다.
- scatter() method는 plot() method와 똑같이 작동한다.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(1024, 2)

plt.scatter(data[:,0], data[:,1])
plt.show()
```

### 막대 차트 그리기
막대 차트를 그리고 싶다. → bar() method를 이용한다.
- bar() method는 x 좌표와 각 막대의 높이를 받는다.
- 수평 막대는 barh()를 통해서 구할 수 있다.
- 막대의 굵기는 width 파라미터로 조정할 수 있다.

```python
import matplotlib.pyplot as plt

data = [5., 25., 50., 20.]

plt.bar(range(len(data)), data)
plt.show()
```

```python
plt.barh(range(len(data)), data)
plt.show()
```

```python
plt.bar(range(len(data)), data, width=0.5)
plt.show()
```

### 다중 막대 차트 그리기.
여러 값을 한 그래프 안에서 비교하고 싶다.
- 구별하기 위한 색깔은 밑에를 고려하면 된다.
- 분할 막대 또한 bottom이라는 파라미터를 통해 구할 수 있다.

![image](https://user-images.githubusercontent.com/61931924/97259071-43981e80-185d-11eb-803a-134faaf69434.png)
![image](https://user-images.githubusercontent.com/61931924/97259125-60cced00-185d-11eb-84bd-45fdb7c2a0ca.png)

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.,],
       [4., 23., 51., 17.],
       [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X + 0, data[0], color='black',width=0.25)
plt.bar(X + 0.25, data[1], color='aquamarine',width=0.25)
plt.bar(X + 0.50, data[2], color='pink',width=0.25)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.,],
       [4., 23., 51., 17.],
       [6., 22., 52., 19.]]

color_list = ['black', 'aquamarine', 'pink']
gap = 0.8 / len(data)
for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i * gap, row, width=gap, color=color_list[i % len(color_list)])
    
plt.show()
```

```python
import matplotlib.pyplot as plt

A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]

X = range(4)

plt.bar(X, A, color = "gold")
plt.bar(X, B, color = "violet", bottom = A)   # A 그래프 밑에 B 그래프가 생긴다.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1., 2., 1., 1.])
X = np.arange(4)

plt.bar(X, A, color = "gold")
plt.bar(X, B, color = "violet", bottom = A)   # A 그래프 밑에 B 그래프가 생긴다.
plt.bar(X, C, color = "forestgreen", bottom = A + B)

plt.show()
```
