### figure와 서브플롯
1. figure 객체를 만들고 싶다. → pyplot의 figure를 사용하자.
- matplotlib에서 그래프는 figure 객체 내에 존재한다. figure라는 객체가 있는 것이다.
- 우선 figure 객체를 생성하면 그 안에 그래프는 아무것도 없다. 앞으로 추가하는 것이다.
- figure 객체를 생성하고 plot method를 통해 plot을 그리면 가장 최근의 서브플롯에 그림이 그려진다.
- 추가 인자: nrows(로우 수), ncols(컬럼 수), sharex(모든 서브플롯이 같은 x축 눈금), sharey(x랑 동일), subplot_kw(

```python
%matplotlib notebook   # 주피터 노트북의 시각화 기능 활성화
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig_ = plt.figure()
```

2. figure 객체의 크기와 비율을 바꾸고 싶다. → figure의 figsize method를 사용하자

```python
fig = plt.figure()
fig.figsize=(10, 5)   # 소괄호 안에 원하는 크기와 비율을 기입하면 된다.
```

3. 그래프를 추가로 작성하고 싶다. → figure의 add_subplot method를 사용하자. 
- 처음 두 칸의 숫자는 가로와 세로의 크기를 의미하며 마지막 숫자는 4개의 서브플롯 중에서 위치를 나타낸다.

```python
fig.add_subplot(2,2,1)
```

```python
fig.add_subplot(2,2,3)
fig.add_subplot(2,2,4)
```

위의 그림을 통해 해당 위치에 만들어진 서브플롯이 들어갔음을 알 수 있다. 

```python
plt.plot(np.random.randn(50).cumsum(), 'k--');
```

4. 서브플롯 간의 간격을 조절하고 싶다. → pyplot의 subplots_adjust method를 사용한다. 
- 원래 간격은 전체 그래프의 높이와 너비에 따라 상대적으로 결정된다. 그래서 굳이 사용할 필요는 없다.

```python
plt.subplots_adjust(wspace=0, hspace=0)   # 다 붙어버리게 만들어버렸다ㅋㅋ
```

### 그래프 저장
그래프를 파일로 저장하고 싶다. → plt.savefig method를 사용하자. 
- 파일 종류는 확장자로 결정된다. png, pdf 등이 있다.
- 인자로는 dpi(인치당 점 해상도, 기본값은 100), bbox_inches('tight'로 지정하면 figure 둘레의 비어있는 공간을 모두 제거), fececolor, edgecolor(바깥 배경 색상)

```python
plt.savefig('figpath.pdf')   # 제목과 확장자만 써주면 되니까 간단하다.
```
