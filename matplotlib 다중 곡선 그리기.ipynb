### 다중 곡선 그리기
곡선을 그리는 이유 중 하나는 곡선을 비교하는 것이다. 더 철저한 조사를 하기 위해서 그래프를 사용할 수도 있다.
- 곡선의 색깔을 정하기 위해서는 color 매개변수를 사용하면 된다.
- 점선으로 표현하고 싶으면 color 매개변수는 사용하지 말고 ""내에 -를 두 번 입력하면 된다. 한 번 입력하면 그냥 곡선.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya, "r-")
plt.plot(X, Yb, 'b--')
plt.show()
```

맷플롯립의 핵심은 지연 렌더링으로 만들 것을 선언하며 이를 맞춰주는 메커니즘이다. _plt.show()_ 명령어를 사용할 때만 그래프를 렌더링한다.  
\*렌더링이란? 컴퓨터 그래픽(스)이나 디지털 애니메이션에서 가상으로 완성된 결과를 만들어 내는 과정. 또는 완성된 화면이나 영상(by. 네이버)

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)

plt.plot(X, Y)
plot_slope(X, Y)
plt.legend(["curve", "slope"])
plt.show()
```

plt.plot()은 렌더링할 것을 선언하고 렌더링 자체에는 아무런 영향을 주지 않으며, 렌더링을 실행시키지 않는다. 많은 곡선이 있는 복잡한 그래프에 대한 스크립트를 작성할 때 매우 유용하다.

### 파일 데이터로부터 곡선 그리기
추가로 파일에서 데이터를 불러와 그래프를 작성해보겠다.

```python
import matplotlib.pyplot as plt

X, Y = [], []   # 상황마다 다르겠지만 이 선언 방식이 되게 편하다고 생각.
for line in open('data/my_data.txt', 'r'):
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])
    
plt.plot(X, Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

with open('data/my_data.txt', 'r') as f:
    X, Y = zip(*[[float(s) for s in line.split()] for line in f])
    
plt.plot(X, Y)
plt.show()
```

### 데이터를 가져오는 코드에서 정밀 검사나 오류 처리를 진행하지 않는 것을 주목해야 한다. 훌륭한 프로그래머는 게으른 프로그래머다.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data/my_data.txt')   # 텍스트 파일을 읽은 후 2D 배열을 반환. 넘파이를 활용한 2차원 배열은 행렬이라 생각하면 된다.

# data는 넘파이 2차원 배열 

plt.plot(data[:,0], data[:,1])
plt.show()
```

대용량 파일인 경우 넘파이를 사용하면 현저하게 빨라진다. 

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/my_data2.txt")
for column in data.T:                 # 데이터 배열의 열로 for문을 수행한다는 의미.
    plt.plot(data[:,0], column)
    
plt.show()
```

data.T는 2차원 배열 데이터의 전치이다. 행과 열을 바꿔 본다.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/my_data2.txt")
for row in data:
    plt.plot(data[0,:], row)
    
plt.show()
```
